```python
# Проверка на наличие NaN в DataFrame
print("=== ПРОВЕРКА НА НАЛИЧИЕ NaN ===")
nan_check = df.isna()
print("Есть ли NaN значения в каждой колонке:")
print(nan_check.any())
print("\nКоличество NaN в каждой колонке:")
print(df.isna().sum())
```

```python
# Заполнение NaN определенным значением (например, нулем)
print("\n=== ЗАПОЛНЕНИЕ NaN ===")
df_filled = df.fillna({
    'Age': 0,
    'Cabin': 'Unknown',
    'Embarked': 'U'  # Unknown
})
print("После заполнения NaN:")
print(f"NaN в Age: {df_filled['Age'].isna().sum()}")
print(f"NaN в Cabin: {df_filled['Cabin'].isna().sum()}")
print(f"NaN в Embarked: {df_filled['Embarked'].isna().sum()}")
```

```python
# Удаление строк, содержащих NaN
print("\n=== УДАЛЕНИЕ СТРОК С NaN ===")
df_dropped = df.dropna()
print(f"Размер исходного датасета: {df.shape}")
print(f"Размер после удаления всех строк с NaN: {df_dropped.shape}")
print(f"Осталось строк: {len(df_dropped)} из {len(df)}")
```

```python
#Часть 4: Задания для практики
#Прочитайте данные из файла 'titanic.csv'. Проверьте, есть ли пропущенные значения в данных и заполните их нулями.
#Выведите первые 10 строк данных. Выберите только те строки, где значение в столбце 'Age' больше 30.
#Отсортируйте данные по столбцу 'Fare' в порядке убывания.
#Сгруппируйте данные по столбцу 'Pclass' и вычислите средний возраст ('Age') для каждого класса.
```

```python
df = pd.read_csv(r'C:\Users\Глеб\Downloads\titanic.csv')

print("Пропущенные значения до заполнения:")
print(df.isnull().sum())

df_filled = df.fillna(0)

print("\nПропущенные значения после заполнения:")
print(df_filled.isnull().sum())

# Выведите первые 10 строк данных
print("\nПервые 10 строк данных:")
print(df_filled.head(10))

age_over_30 = df_filled[df_filled['Age'] > 30]
print(f"\nКоличество пассажиров старше 30 лет: {len(age_over_30)}")

sorted_by_fare = df_filled.sort_values('Fare', ascending=False)
print("\nТоп-5 самых дорогих билетов:")
print(sorted_by_fare[['Name', 'Pclass', 'Fare']].head(5))

average_age_by_class = df_filled.groupby('Pclass')['Age'].mean()
print("\nСредний возраст по классам:")
print(average_age_by_class)
```
