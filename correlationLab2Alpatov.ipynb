```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
menDf = brainFrame[brainFrame['Gender'] == 'Male']
womenDf = brainFrame[brainFrame['Gender'] == 'Female']
```

```python
# Ячейка для кода № 6
menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
# Ячейка для кода № 7
# Постройка графика диаграммы рассеяния для кадра данных с женскими записями
womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
brainFrame.select_dtypes(include=[np.number]).corr(method='pearson')
```

```python
womenDf.select_dtypes(include=[np.number]).corr(method='pearson')
```

```python
menDf.select_dtypes(include=[np.number]).corr(method='pearson')
```

```python
#Диагональ заполнена значениями 1, потому что это корреляция переменной с самой собой. Коэффициент корреляции Пирсона между одной и той же переменной всегда равен 1, так как они идеально связаны линейной зависимостью.
```

```python
#Значения зеркалируются потому, что корреляционная матрица симметрична. Корреляция между переменной A и B такая же, как между переменной B и A (corr(A,B) = corr(B,A)), поэтому матрица симметрична относительно главной диагонали.
```

```python
numeric_columns = ['PIQ', 'FSIQ', 'VIQ', 'Weight', 'Height', 'MRI_Count']
```

```python
# Ячейка для кода № 12
import seaborn as sns

wcorr = womenDf[numeric_columns].corr()
sns.heatmap(wcorr)
plt.show()
#plt.savefig('attribute_correlations.png', tight_layout=True)
```

```python
# Ячейка для кода № 14
mcorr = menDf[numeric_columns].corr()
sns.heatmap(mcorr)
plt.show()
#plt.savefig('attribute_correlations.png', tight_layout=True)
```

```python
#Корреляция близкая к нулю означает, что между этими переменными нет линейной зависимости. Изменение одной переменной не связано с изменением другой. Например, если корреляция между ростом и IQ близка к нулю, это значит, что высокие люди не обязательно умнее низких и наоборот.
```

```python
#Разделение по полу необходимо потому, что:
#Мужчины и женщины имеют разные физиологические особенности (средний рост, вес, размер мозга)
#Без разделения можно получить ложные корреляции из-за смешивания двух разных групп
#Это позволяет выявить полоспецифичные закономерности
```

```python
#Скорее всего, наиболее сильную корреляцию с размером мозга (MRI_Count) будут иметь:
#Рост (Height) и Вес (Weight) - так как размер мозга коррелирует с общими размерами тела
#Показатели интеллекта (PIQ, FSIQ, VIQ) - могут иметь слабую положительную корреляцию
#Это ожидаемо, поскольку:
#Большие люди обычно имеют больший мозг
#Связь между размером мозга и интеллектом является предметом научных дискуссий, но обычно довольно слабая
#Биологические параметры (рост, вес) обычно сильнее коррелируют с физическими характеристиками, чем когнитивные
```
